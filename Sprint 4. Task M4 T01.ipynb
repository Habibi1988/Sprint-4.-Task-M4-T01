```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('preu lloguer per trimestre i barri.csv') # Original data frame.

df
```

```python
df.isnull().sum()
```

```python
# I filtered the data frame column 'Lloguer mitjà'
#  and now just (Lloguer mitjà mensual (Euros/mes)) shows.

filtered_df= df[df['Lloguer_mitja'].str.contains ('(Lloguer mitjà mensual)')] 

filtered_df
```

```python
filtered_df.info() # I want to change the 'preu' column to foat.
```

```python
int_df= filtered_df.copy()
int_df['Preu'] = int_df['Preu'].replace('--', np.nan) 
int_df['Preu'] = int_df['Preu'].astype(float) # changed to float.
```

```python
int_df.info()
```

```python
int_df.isnull().sum() # I want to replace the  null values with average by district.
```

```python
# I replaced null values of 'preu' column with average by district.
# Calculate district averages
cleaned_df=int_df.copy()
district_averages = cleaned_df.groupby('Nom_Districte')['Preu'].mean()

# Create a mapping of districts to average prices
district_mapping = district_averages.to_dict()

# Impute missing values based on district averages
for index, row in cleaned_df.iterrows():
    if pd.isnull(row['Preu']):
        district = row['Nom_Districte']
        cleaned_df.at[index, 'Preu'] = district_mapping[district]
cleaned_df
```

```python
cleaned_df.describe().T
```

```python
cleaned_df.isnull().sum()
```

### Create at least one view for:
- A categorical variable (District or Neighborhood)
- A numeric variable (Price)
- A numerical variable and a categorical one (District and Price)
- Two numeric variables (Year or Quarter and Price)
- Three variables (Neighbourhood or District, Quarter and Price)

```python
# A categorical variable (District) 
cleaned_df['Nom_Districte'].value_counts().plot(kind='bar')
plt.title('just district')
plt.show()
```

```python
cleaned_df['Preu'].describe()
```

```python
# A numeric variable (Price)
sns.boxplot(data=cleaned_df, x='Preu')
plt.show()
```

```python
sns.barplot(data=cleaned_df, x='Preu', y= 'Nom_Districte')
plt.title('price and district')
plt.show()
```

```python
# I don't like to sort the price column but for better apearence, in this level I sort the price.
sorted_df= cleaned_df.copy()
sorted_df.sort_values(by= 'Preu', ascending = False, inplace= True)
```

```python
# once again I want to repeate this with sorted values.
sns.barplot(data=sorted_df, x='Preu', y= 'Nom_Districte')
plt.title('price and district in another way')
plt.show()
```

```python
# with box plot.
sns.boxplot(data=sorted_df, x='Preu', y= 'Nom_Districte')
plt.title('price and district with boxplot')
plt.show()
```

```python
# Two numeric variables (Year or Quarter and Price)

sns.histplot(data=sorted_df, x= 'Preu', y= 'Trimestre' )
plt.title('quarter and price')

plt.show()
```

```python
# Three variables (Neighbourhood or District, Quarter and Price)

plt.figure(figsize=(15, 4))
sns.barplot(x='Trimestre', y='Preu', hue='Nom_Districte', data=sorted_df)
plt.xlabel('Trimestre')
plt.ylabel('Preu')
plt.title('Quarterly Prices by district')
plt.legend(title='district', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```
